```python
from ultralytics import YOLO
import torch
import numpy as np

import easyocr
import cv2
import matplotlib.pyplot as plt
```

```python
PATH_TO_DATASET = "../Master-Documents-OCR/datasets"
if torch.backends.mps.is_available():
    device = torch.device("mps")

model = YOLO('yolov8n.yaml')

results = model.train(data=f'{PATH_TO_DATASET}/data.yaml', epochs=30, imgsz=640, device=device)
```

```python
class Detector:
    def __init__(self, path_to_weights: str = 'weights/detector.pt', confidence_level: float = 0.25):
        self.model = YOLO(path_to_weights)
        self.model.conf = confidence_level
        # self.names = {0: 'number', 1: 'series'}

    def predict(self, image: np.array):
        outputs = self.model.predict(image, task="detection", max_det=4)[0]
        # output_image = outputs.orig_img
        names = outputs.names

        predictions = []
        for output in outputs.boxes:
            x1, y1, x2, y2 = output.xyxy[0].numpy()
            predictions.append(
                {
                    'class': output.cls.int().item(),
                    'name': names[output.cls.int().item()],
                    'confidence_level': output.conf.item(),
                    'xyxy': output.xyxy.numpy(),
                    'image': image[int(y1):int(y2), int(x1):int(x2)]
                } 
            )
        
        return predictions
```

```python
class Recognitor:
    def __init__(self, ):
        self.model = easyocr.Reader(['ru', 'en'], gpu=False)

    def predict(self, image: np.array):
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        predictions = self.model.recognize(gray_image, rotation_info=[0, 90, 180, 270])[0]

        return predictions[1] # Ð¢Ð¾Ð»ÑŒÐºÐ¾ Ñ‚ÐµÐºÑÑ‚
```
